```python
import seaborn as sns
df=sns.load_dataset('car_crashes')
df
```

####Scatterplot

```python
sns.scatterplot(x="total",y="alcohol",data=df)
```

INFERENCE: alcohol and total are directly proportional. with increase in alcohol, total also increases.

####Lineplot

```python
sns.lineplot(x="alcohol",y="no_previous",data=df,ci=None)
```

INFERENCE: with increase in alcohol, no_previous also increases

####Distribution plot

```python
sns.distplot(df["alcohol"])
```

INFERENCE: The graph represents overall distribution on continuous data variable alcohol against its density distribution using a histogram and line graph

####Relation plot

```python
sns.relplot(x="alcohol",y="not_distracted",data=df,hue="abbrev")
```

INFERENCE: every element in column abbrev is given a different colour and plotted against alcohol and not_distracted deoending upon their occurences.

```python
df["abbrev"].value_counts()
```

####Barplot

```python
sns.barplot(data=df,x="abbrev",y="alcohol",ci=None)
```

INFERENCE: plots the level of alcohol based on the abbrev.

```python
sns.barplot(data=df,x="alcohol",y="speeding",hue="abbrev")
```

####Countplot

```python
sns.countplot(x="alcohol",data=df)
```

INFERENCE: compares the count across nested variables, showing the number of variables with each value of alcohol

####Jointplot

```python
sns.jointplot(x="speeding",y="alcohol",data=df)
```

INFERENCE: we observe both the scatterplot and the barplot, there is a positive relationship between alcohol and speeding as one increases with the other. Skewness is symmetric as the barplot is neither tending towards right or left. points lying far away from the scatterplot are the outliers.

####Boxplot

```python
sns.boxplot(x="speeding",y="ins_premium",data=df)
```

INFERENCE: boxplot representing upper and lower quartile range, in this case skewness is symmetric.

###Correlation

```python
cor=df.corr()
cor
```
